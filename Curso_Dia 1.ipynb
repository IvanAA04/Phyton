### Ingresar datos por parte del usuario, sin necesidad de declarar el valor junto con la variable

### Operaciones Matemáticas

```python
# 4to Nivel de importancia
#Suma ordinaria
print("suma ordinaria")
print(3 + 4)                  # 7
variable1,variable2 = 3,4     # 60
print(variable1 + variable2)

# Resta ordinaria
print("resta ordinaria")
print(2 - 5)

# Declaración de variables
v1 = 10.4
v2 = 15
resultado = v2 - v1
print("resta declarada en una variable")
print(resultado)  #4.6
```

```python
# 3er nivel de importancia
# Multiplicación ordinaria *
print("Multiplicación ordinaria")
M1, M2 = 2,3.14
print(M1 * M2)

# División ordinaria /
print("----------------------")
print("División ordinaria")
M1, M2, M3, M4, M5, M6 = 2,7,6,4,8,3
print(M1 / M2)
print(M3 / M4)
print(M5 / M6)

# División de piso //
print("----------------------")
#los resultados arrojados son redondeandos hacia abajo para asegurar que siempre devuelva un entero
print("División de piso")
print(M1 // M2)
print(M3 // M4)
print(M5 // M6)
```

```python
# Potencias **
print("Potencias")
print(9 ** 1/3)  # Sacar raíz de 9
P1,P2 = 7,2
print(P1 ** P2)
```

```python
# Área de un circulo
pi = 3.14
r = 6
A = pi * (r ** 2)
print(A)
```
