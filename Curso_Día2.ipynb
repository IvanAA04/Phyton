## Elif

Ahora hagamos un programa en el que preguntamos la calificación del 0 al 100 y queremos saber su equivalente en el sistema de evaluación estadounidense. Para simplicidad, basemonos en esto:

- A+ = 100
- A = 90 - 99
- B = 80 - 89
- C = 70 - 79
- D = 60 - 69
- F < 60



```python
cal = 90
if cal == 100:
  print("A+")
elif cal >= 90:  # El elif te permite hacer lo mismo que el if, pero es una condición que se evalua primero el if y luego al 2ndo if, en este caso es el elif
  print("A")
elif cal >= 80:
    print("B")
elif cal >= 70:
    print("C")
elif cal >= 60:
    print("D")
else:   # Se emplea un else ya que por descarte si no es una de las demás opciones debe ser una F
    print("F")
```

# Operadores lógicos con Condicionales

```python
# and, or, not
pan = True
queso = True
mayonesa = True
jamon = False
print(pan and queso and mayonesa and jamon)
if not (pan and queso and mayonesa and jamon):  # Debe ser False o 0 para que la condición not se active y pase lo del if
  print("No se puede hacer el sandwich")
else:
  print("Se puede hacer el sandwich")
```

```python
cal = 90
if cal >= 100 and cal <= 100:
  if cal == 100:
    print("A+")
  elif cal >= 90:  # El elif te permite hacer lo mismo que el if, pero es una condición que se evalua primero el if y luego al 2ndo if, en este caso es el elif
    print("A")
  elif cal >= 80:
      print("B")
  elif cal >= 70:
      print("C")
  elif cal >= 60:
      print("D")
  else:   # Se emplea un else ya que por descarte si no es una de las demás opciones debe ser una F
      print("F")
else
  print("Calificación no válida, debe estar entre 0 y 100")
```

```python
# Módulo (%)
# Te devuelve el residuo de una división, es decir, lo que sobra(decimales)
# 10 % 3 da 1, porque 10 dividido entre 3 cabe 3 veces y sobra 1
# 7 % 2 da 1, porque el 7 no es divisible de forma exacta entre 2
print(10 % 3)
print(7 % 2)

num = 9
if num % 3 == 0 or num % 5 == 0:
  print(f"El numero {num} es multiplo de 3 o de 5")

# saber si el numero es par o impar usando not
num = 6
if num % 2 == 0:
  print(f"El numero {num} es par")
else:
  print(f"El numero {num} es impar")

```

## Ciclo For
los llamados ciclos for nos ayudan a realizar acciones durante una cantidad determinada de veces, es decir, ya se conoce la cantidad de veces que se va a ejecutar.

```python
# Imprimir los numeros del 0 al 4
for i in range(5): #Toma 5 como el numero final pero inicia SIEMPRE en 0
  print(i) #la variable i solo existe mientras se emplea el for
```

```python
for num in range(2,11): # inicia del 2 al 10, no hasta el 11 porque al imprimir el 2 ya uso un numero y le quedan otros 10 ciclos donde suma 1
  print(num)
```

```python
# También existe el range(inicio, fin, paso)
for i in range(2,21,2): # te imprime los numeros desde el 2 al 21 y te los imprime de 2 en 2.
  print(i)
```

```python
cadena = "Python"
for letra_por_letra in cadena:
  print(letra_por_letra)
```

# While

```python
suma = 0
numeros = int(input("Ingresa un numero para sumarlo o ingresa un 0 para salir: "))
while numeros != 0:  # siempre y cuando el numero sea diferente de 0, el ciclo continuará
  suma += numeros
  numeros = int(input("Ingresa un numeropara sumarlo o ingresa un 0 para salir: "))

print(f"Tu suma es: {suma}")
```

```python
# Emplear el while como for
numero = 0
while numero <= 12:
  print(numero)
  numero += 1
```

```python
print("Juego: Adivina el Número")
numero_secreto = int(input("Jugador 1, escribe el número secreto (1-20): "))
intentos = 0
max_intentos = 4
while intentos < max_intentos:
  intento = int(input("Adivina el número: "))
  intentos += 1

  if intento == numero_secreto:
    print("¡Correcto!")
    break

  else:
    print("Incorrecto, intenta de nuevo.")
    #Opcional
    if intentos == max_intentos and intento != numero_secreto:
      print("Se acabaron los intentos.")
      print("El número era:", numero_secreto)
```
