# Estructura de datos
Las estructuras de datos son formas de organizar la información para trabajar con ella de manera eficiente

- Listas -> ordenadas, modificadas, permiten duplicados
- Tuplas -> ordenadas, inmutables, permiten duplicados
- Sets -> no ordenadas, sin duplicados
- Diccionarios -> pares clave:valor

Nota: Los siguientes son estructuras que no se veran en el curso, pero existen y tienen sus respectivos usos

- Pilas (FILO -> First In, Last Out) -> el primer dato que entra es el último que sale
- Colas/filas (FIFO -> First in, First out) -> el primer dato que entra es el ultimo que sale
- Listas enlazadas (o doblemente enlazadas)


## Listas []
Son colecciones ordenadas que podemos modificar y almacenan elementos de distintos tipos de datos

Son muy parecidas a los arreglos


```python
# Ejemplo de un arreglo que funciona como una lista de compras
# Estos se van agregando ocupando espacios enumerados del 0 hasta la cantidad de datos que haya
lista_de_compras = ["manzanas", "pan", "leche", "huevos", "mantequilla", "queso", "jamon"]
print(lista_de_compras)

# para acceder a uno de los datos dependiendo de la posición que ocupa
print(lista_de_compras[6])
```

```python
# Al agregar un elemento, este de manera automatica se agrega al final SI no se específica
# Agregar un elemento a la lista .append()
lista_de_compras.append("sopa")
print(lista_de_compras)
```

```python
# Para agregar un elemento en una posición específica
lista_de_compras.insert(1,"platano") # sintaxis -> arreglo.instert(posición, dato)
print(lista_de_compras)
```

```python
# Para eliminar un dato en específico
lista_de_compras.remove("manzanas") # sintaxis -> arreglo.remove(dato)
print(lista_de_compras)
```

```python
# Para eliminar un elemento en una posición específica

lista_de_compras.pop(1) # sintaxis -> arreglo.pop(posición)
lista_de_compras.pop() # Sin nada dentro del espacio de posición borrará el ultimo elemento
print(lista_de_compras)

# Nota: Te retorna el numero eliminado
```

```python
#
va = "Python"
va[3]
```

```python
# Podemos recorrer el arreglo por medio de un for
for elementos in lista_de_compras: # la variable "elementos" permite recorrer todos los datos que el arreglo
    print(elementos)
```

```python
# Para sacar el tamaño total de la lista
# len() saca el tamaño de la lista
print(len(lista_de_compras)) # cantidad de elementos dentro
for i in range(len(lista_de_compras)):
    print(f"{i + 1} - {lista_de_compras[i]}") # se le suma un 1 porque el primer espacio del arreglo es 0
```

```python
promedios = [55.6,97.5,23.6,97.5,72.4,100]
print(promedios)
promedios.sort() # sort() sirve para ordenar la lista de manera ascendente
print(promedios)
 # ordenar la lista de manera descendente
promedios.sort(reverse=True)
print(promedios)

```

```python
# Podemos ingresar al ultimo elemento de una lista
print(promedios[-1])
```

```python
# Encontrar el mayor y el menos de una lista
print(max(promedios))
print(min(promedios))
```

```python
# Declaración y uso de matrices
calificaciones = [[100,88,93], [99,56,72]] # En el primer corchete esta el primer usuario y sus calificaciones por materia, el 2ndo corchete es el 2ndo usuario y sus calificaciones.
print(calificaciones)   # la matriz es de 2x2
i = 1
for cali_1 in calificaciones:
    print(f"\n Alumno {i}")
    j = 1
    for cali_j in cali_1:
        print(f"Cali {j}:  {cali_j}")
        j += 1  # Al termino de cada ciclo se le sumo 1 a la variable por cada dato que hay

```

## Tuplas ()
Colecciones ordenadas pero inmutables.



```python
# Hagamos un programa donde guardemos fechas de nacimientos de un grupo en una tupla
fechas = ("10/10/05","22/03/02","17/04/04", "25/02/04")
print(fechas)
print(type(fechas))
```

## SETS {}

Los sets son colecciones SIN orden y que NO permiten elementos duplicados

```python
nombre = {"Rogelio", "Jorge", "Jorge", "Francisco"}
print(nombre)
print(type(nombre))
```

```python
# puede almacenar diferentes tipos de datos
distintos_datos = {"Angel", 1, "Diana", "Juan", 4.8}
print(distintos_datos)
```

```python
# Agregar datos al set, cabe mencionar que esto es al azar debido a las propiedades del set
# Add
nombre.add("Leonardo")
print(nombre)
```

```python
# Para demostrar que no tienen orden se va a tratar de imprimir alguna posición especifica
print(nombre[0])
```

```python
# Tampoco puedes agregar nombres en alguna posición específica
nombre[2] = "Rogelio"
```

```python
# Unión
matutino = {20322,22183,12356,56485}
vespertino = {20322,22183,56456,54541}
# union empleando .union()
matriculas = matutino.union(vespertino)
print(matriculas)
# union empleando |
total_matriculas = matutino | vespertino
print(total_matriculas)
```

```python
# Intersección
ca = {1234,211,574,126,584,693,648,6526,135,12258,695}
al = {1234,211,574,126,584,693,513,5465,5448,1283,145}

# intersección con .intersection()
ca_al = ca.intersection(al)
print(ca_al)
# intersección empleando &
ca_al = ca & al
print(ca_al)
```

```python
# Diferencia
# Diferencia usando .difference
solo_calculo = ca.difference(al)
print(solo_calculo)
# Diferencia usando -
solo_calculo = ca - al
print(solo_calculo)
```

```python
#
nombre = {"Rogelio", "Jorge", "Jorge", "Francisco"}

# remove, pop
nombre.remove("Rogelio") #con remove debes de específicar cual dato quieres eliminar
print(nombre)
nombre.pop() # El pop elimina uno de manera aleatoria
print(nombre)
```

```python
# vaciar todo el set
nombre.clear()
print(nombre)
```
