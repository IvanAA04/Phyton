## Funciones
Las funciones son bloques de código reutilizable que podemos mandar a llamar varias veces en el código, de tal manera que no sea necesario poner la parte de código, sino simplemente la función.

```python
def mensaje():
  print("Hola!!!")

mensaje()
```

```python
# Función con retorno
def mensaje():
  return "Hola!!!"

print(mensaje())
va = mensaje() # También puedes guardar lo que retorna una función en una variable
print(va)
```

```python
def saludo(nombre): # p2 En esta parte se carga y recibe la variable enviada en p1 y se interpreta para luego ser retornada
  print("Hola, ", nombre, "!!!")  # p3 Esta parte es el conjunto de instrucciones y en la que imprime el mensaje con tu nombre
tuNombre = input("Ingresa tu nombre: ")
saludo(tuNombre) # p1 En esta parte se la da la función saludo(nombre) el dato de la variable llamada "tuNombre"
# Nota: el nombre de la variable del parametro puede o no ser el mismo fuera de la función
```

```python
# Puedes mandarle el parametro de esta manera
saludo("Grupo")
```

```python
# Puedes hacer que tus variables trabajen con dos parametros
# en lugar de uno, de tal manera que pueda procesar más información

def suma(numero_1,numero_2):
  return numero_1 + numero_2
print(suma(2,5)) # Puedes directamente imprimir la función dandole los valores por default

suma = suma(1,2) # Se puede almacenar el valor retornado dentro de una variable
print(suma) # y luego imprimirla

# Nota: si cambias de lugar el orden de las impresiones obtendras un error ya que estas sobreescribiendo datos
```

```python
def bienvenida(nombre = "invitado"): # En esta parte se le asigna un valor inicial a la variable en caso de que el usuario no ponga nada
  print("Hola", nombre)
bienvenida() # 1era impresión
bienvenida("Alex") # Dato que se le entrega a la variable de nombre
```

```python
def suma(numero_1 = 2, numero_2 = 10): # Es posible asignar multiples variables
  return numero_1 + numero_2
suma = suma(100000, 23143) # NO se puede acceder al 2ndo valor si no mandas un primero, si quitas el 1er valor no podrás asignar un 2ndo
print(suma)
```

```python
# Ejercicio - asignar un descuento

def descuento(precio, desc = 0.15): # la 2nda variable de desc es la que tiene el porcentaje de descuento aplicado
  precio = precio * (1 - desc)
  print("El nuevo precio es: ",precio)

descuento(750) # Precio del articulo con el descuento por default
descuento(800,.25) # Imaginemos que uno de los articulos recibe un descuento mejor
```

```python
def saludar(nombre):
  return f"hola, {nombre}!"

def despedirse(nombre):
  return f"Adios, {nombre}!"

def mensaje_completo(nombre):
  saludo = saludar(nombre)
  despedida = despedirse(nombre)
# return saludo + "\n" + despedida   # Esta es otra forma de escribirlo
  return f"{saludo}\n{despedida}"

print(mensaje_completo("Alex"))
```

```python
# Ahora supongamos que tenemos una heladeria
def pedir_sabor():
  sabores = ["vainilla", "chocolate", "fresa"] # [] de lista
  print("Sabores disponibles", ",", sabores)
  # lower() hace minusculas las palabras
  # strip() elimina los espacios tanto antes como despues
  sabor = input("Elige un sabor: ").strip().lower()
  # en el while de abajo nos aseguramos de que el usuario escoja un sabor valido
  while sabor not in sabores:
    sabor = input("Elige un sabor (nuevamente): ").strip().lower()
  return sabor

def pedir_cantidad():
  cantidad = int(input("Cuántas bolas de nieve? "))
  while cantidad <= 0:
    cantidad = int(input("Cuántas bolas de nieve? (ingresa un valor válido): "))
  return cantidad

# Función madre (Llamando a esta función llamamos a todas las anteriores)
def pedir_helado():
  sabor = pedir_sabor()
  cant  = pedir_cantidad()
  print(f"Pedido: {cant} bola(s) de {sabor}")
  return cant, sabor

# Ya aquí solo se ejecuta el codigo dependiendo de la cantidad de clientes a atender
total_pedidos = int(input("Cuántos clientes atenderas? "))
for i in range(total_pedidos):
  print(pedir_helado())
```

## Funciones Lambda
Una función lambda es una función acortada para escribir una función de esta forma:

nombreFunción = lambda parametros: accion

La diferencia entre las funciones lambda y las funciones normales es que suelen ser un poco más limitadas

```python
sumar = lambda x,y : x + y
print(sumar(2,3))
```

```python
descuentoLambda = lambda desc, precio: ((precio * (1 - desc)), (precio * desc))
nuevoPrecio = descuentoLambda(0.15, 1000)
print("El precio a pagar es: ", nuevoPrecio[0], ", el descuento que se le hizo fue de: ", nuevoPrecio[1])
print(type(nuevoPrecio)) # Retorna una tupla
```

```python
# Calculadora -> suma,resta, producto, división y potencia
def suma(a,b):
  return a + b
def resta(a,b):
  return a - b
def producto(a,b):
  return a * b
def division(a,b):
  if b == 0:
    print("Error: el divisor no debe de ser 0")
    return
  return a / b
def potencia(a,b):
  return a ** b

def menu():
  print("-------Calculadora--------")
  print("1. Suma")
  print("2. Resta")
  print("3. Producto")
  print("4. División")
  print("5. Potencia")
  print("6. Salir")

while True:
  menu()
  try:
    op = int(input("Seleccione una opción: "))
    if op < 1 or op > 6:
      print("Opción invalida")

  except ValueError:
    print("Ingrese un numero valido")
    if op == 6:
      print("Saliendo...")
      break

  try:
    a = float(input("Ingrese el primer numero: "))
    b = float(input("Ingrese el segundo numero: "))
  except ValueError:
    print("Error: Ingrese un numero valido")
  if op == 1:
    print("Resultado: ", suma(a,b))
  if op == 2:
    print("Resultado: ", resta(a,b))
  if op == 3:
    print("Resultado: ", producto(a,b))
  if op == 4:
    print("Resultado: ", division(a,b))
  if op == 5:
    print("Resultado: ", potencia(a,b))
  else:
    print("Opcion invalida")
```
