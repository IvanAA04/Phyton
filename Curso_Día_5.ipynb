## Qué es un módulo?

Un módulo es una unidad de organización del código que permite encapsular y dividir el software en partes más manejables, generalmente representado como un archivo que contiene código relacionadom, como funciones, variables (constantes) o métodos de clase.

```python
# Algunas librerias vienen por defecto en phyton
import math

r = math.sqrt(2) # es una forma de llamar la función sqrt de la librebria math y obten la raíz cuadrada de 2
print(r)

print(math.pi) # imprimir pi
area = math.pi * math.pow(2, 2)
print(f"r = {r:.4f}, área = {area:.2f}") # Se imprime el resultado solo con dos decimales
```

```python
# Libreria Random
# Generación de numero pseudoaleatorios y muestreo ()
import random
nums = [random.randint(1,100) for _ in range(5) ] # Cuando se deja el _ singifica que no se usa la variable "i"
# Es como decirle: Traeme en esta lista 5 valores aleatorios entre 1 y 100
print(nums)
```

```python
# Hagamos una contraseña aleatoria de longitud 12
import random
import string

# el string.ascii_letters trae todas las letras de la a a la z tanto en mínusculas como máyusculas
# string.digits trae todas los numeros del 1 al 9
# string.punctuation trae signos de puntuación
todos = string.ascii_letters + string.digits + string.punctuation
print(todos) # Imprimimos para ver los caracteres que nos trae
print(type(todos))
longitud = 12
password = "".join(random.choice(todos) for _ in range(longitud))
print(password)
```

# Existe 1 solo tema que TODO programador independientemente del area DEBE de saber

JSON es un formato para guardar y transmitir datos. Es muy empleando en APIs, archivos de configuración y almacenamiento de información.
(Se parece a un diccionario de Python, pero solo texto)

Para que sirve?
- Sirve para convertir objetos de python -> texto JSON y también texto JSON -> objetis de Python

Esto se llama:
- Serializar: Convertir un objeto de Python en texto JSON
- Deserializar: Convertir un texto JSON a un objeto usable de Python


```python
# Funciones principales
#1) json.dums(obj): Convierte un diccionario u objeto de Python en un string JSON
import json
obj = {"Nombre": "Ana"
, "Edad": 25}
print(type(obj))
s = json.dumps(obj, ensure_ascii = False)
print(s)
print(type(s))
```

```python
# Esta función de abajo convierte strings de JSON para que podamos tratarlas igual que los diccionarios de pyton ()
# json.loads(texto_json): Convierte un string JSON en un diccionario python
restaurar = json.loads(s)
print(restaurar)
print(type(restaurar))
# SIEMPRE debes de asegurarte de que API vienen los datos ya que pueden venir erroneos
# Al venir de una pagina/persona/archivo desconocid@ debe tener un cuidado

'''
Pueden contener:
Datos malinformados
Información inesperada
Campos faltantes
Datos demasiado grandes
variables incorrectas que rompan tu programa
'''
```

```python
'''#UTF es la abreviatura de "8-bit Unicode Transformating Format" y
se trata de un formato de codificación de caracteres Unicode e ISO 10646
que utiliza símbolos de longitud variable, permitiendo representar cualquier
carácter Unicode con una eficiencia que lo hace ideal para internet.'''

with open("datos.json", "w", encoding="utf-8") as f:
    json.dump(obj, f, ensure_ascii=False, indent=4)
```

```python
with open("datos.json", "r", encoding="utf-8") as f:
  # lo ideal es hacer: variable = json.load(f)
  print(json.load(f))
```

## Numpy
Provee el tipo de dato ndarray, operaciones vectorizados, broadcasting, utilidades de algebra lineal y funciones matemáticas en arrays. Es la base sobre la que se apoyan muchas librerias cientificas en Python

### Nota IMPORTANTE: El arreglo de Numpy no es igual a la lista de Python, este NO puede tener diferentes tipos de datos en el mismo

```python
#pip install numpy <- en caso de que no este instalada simplemente se pone el comando
import numpy as np # Se cambia el nombre de la libreria a np
A = np.array([[1,2,3],[4,5,6]]) # Es un arreglo de dos cortes array(2,3)
v = np.array([10,20,30]) #array (3)
print(A + v)
print("--------------")
prueba = [1,2,3]
prueba2 = [1,2,3]
print(prueba + prueba2)
```

```python
# Más funciones de Numpy
z =np.zeros((3,3)) # array(3,3) # Matriz 3x3 llena de 0
print(z)
o = np.ones((3,3)) # matris 3x3 llena de 1
print(o)
r = np.random.rand(3,3) # Matriz 3x3 llena de numeros aleatorios entre 0 y 1
print(r)
```

```python
# Rangos numericos
nums = np.arange(0,10,2) # saca los valores entre 0 y 10 de 2 en 2
print(nums)
x = np.linspace(0,1,5) # saca 5 valores espaciados de manera "equitativa" entre 0 y 1
print(x)
```

```python
a = np.array([1,2,3,4,5])
b = np.array([6,7,8,9,10])
print(a + b) # Suma
print(a - b) # Resta
print(a * b) # Multiplicación
print(a / b) # División
print(a ** b) # Potencia
```

```python
print(a.dot(b)) # Producto entre dos matrices
c = np.array([[1,2,3],[4,5,6]])
d = np.array([[10,20],[30,40],[50,60]])
#1 *10 + 2 * 30 + 3 * 50
#
#
#
print(c.dot(d))
```

```python
data = np.array([1,2,3])

print(data.min()) # Mínimo
print(data.max()) # Máximo
print(data.sum()) # Suma
print(data.mean()) # Promedio
print(data.std()) # Desviación estandar
```

```python
A = np.array([1,6,3,9,2])
print(A[A > 5])
```
